```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
```

```python
racoon= scipy.misc.face()
plt.figure(figsize=(10, 3.6))
plt.imshow(racoon, cmap=plt.cm.gray)
plt.show()
```

```python
grayimg= scipy.misc.face(gray=True)
plt.figure(figsize=(10, 3.6))
plt.imshow(grayimg, cmap=plt.cm.gray)
plt.show()
```

```python
convertedimg= grayimg.reshape(grayimg.shape[0]*grayimg.shape[1],1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(convertedimg)
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(grayimg.shape[0],grayimg.shape[1]) 
```

```python
plt.figure(figsize=(10, 3.6))
plt.imshow(labels)
plt.show()
```

```python
plt.figure(figsize=(10, 3.6))
plt.imshow(labels, cmap=plt.cm.gray)
plt.show()
```
