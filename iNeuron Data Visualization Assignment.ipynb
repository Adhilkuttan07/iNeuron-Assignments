```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("C:\\Users\\Adhil E\\Desktop\\Iris.csv")
df.drop('Id',axis=1,inplace=True)
```

```python
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
```

```python
x=df.iloc[:,:4]
y=df.iloc[:,4]
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
x=principalComponents = pca.fit_transform(x)
col=['Dimension1','Dimension2','Dimension3']
x=pd.DataFrame(data=x,columns=col)
newdf=pd.concat([x,y],axis=1)
```

```python
fig = plt.figure()
ax = Axes3D(fig)

x_vals = list(newdf['Dimension1'])
y_vals = list(newdf['Dimension2'])
z_vals = list(newdf['Dimension3'])
c_vals = list(newdf['Species'])


ax.scatter(x_vals,y_vals,z_vals,c=c_vals)
plt.show()
```
